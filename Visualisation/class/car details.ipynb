```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO data set.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='seller_type', data=df, palette='viridis')
plt.title('selling_type count plot')
plt.xlabel('seller type')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['year'].dropna(),bins=30,kde=True, color='green')
plt.title('Hisplot for year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
cars=df['owner'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owneer', 'Second Owner', 'Third Owner', 'Fourth & Above Owner', 'Test Drive Car'],
        autopct='%1.1f%%', colors=['lightcoral','green','red','green','royalblue' ],
        startangle=180, explode=(0.05,0.05,0.05,0.05,0.05))
plt.title('Cars Details')
plt.show()
```

```python
cars=df['owner'].value_counts()
cars
```

```python
plt.Figure(figsize=(8,6))
sns.boxplot(x='seller_type',y='selling_price',data=df, palette='Set2')
plt.title('Car Details ')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='year', y='km_driven', hue='owner',
                style='seller_type',data=df, palette='rainbow')
plt.title('year vs km_driven with owner and seller_type')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.legend(title='about Car')
plt.show()
```

```python
df.info()
```

```python
num=df[['year','selling_price','km_driven']].corr()
num
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num,annot=True, cmap='rainbow',center=0)
plt.title('correlation Heatmap of Car Dataset')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='fuel', y='year', hue='seller_type', data=df,
               split=True, palette='Set2')
plt.title('seller_type Distribution by year and fuel')
plt.xlabel('fuel')
plt.ylabel('year')
plt.legend(title='Survived')
plt.show()
```

```python
sns.kdeplot(data=df['year'].dropna(),color='skyblue',shade=True, alpha=0.5)
plt.title('year Distribution(Kernal Density Estimate)')
plt.xlabel('year')
plt.ylabel('Density')
plt.show()
```

```python
g= sns.FacetGrid(df, col='seller_type', row='owner', margin_titles=True)
g.map(sns.histplot,'year',bins=20,color='teal')
g.fig.suptitle('age distribution by class and syrvival',y=1.03)
plt.show()
```

```python
sns.pairplot(df[['year', 'selling_price', 'km_driven', 'transmission']],
             hue='transmission', palette='husl',
             plot_kws={'alpha':0.6}, diag_kind='kde')
plt.suptitle('car details')
plt.show()
```

```python
plt.Figure(figsize=(5,5))
sns.swarmplot(x='fuel', y='km_driven', hue='transmission',
              data=df, palette='Set1', dodge=True)
plt.title('transmission Distribution by fuel and km_driven (Swarm Plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```
