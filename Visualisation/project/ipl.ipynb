```python
plt.figure(figsize=(8,6))
sns.countplot(x='batting_team', data=df, palette='viridis')
plt.title('ipl batting team count')
plt.xlabel('batting team')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['batting_team'].dropna(),bins=30,kde=True, color='green')
plt.title('histplot of team ')
plt.xlabel('batting')
plt.ylabel('count')
plt.show()
```

```python
cars=df['batting_team'].value_counts().head(5)
plt.figure(figsize=(8,6))
plt.pie(cars, labels=['First Owneer', 'Second Owner', 'Third Owner', 'Fourth & Above Owner', 'Test Drive Car'],
        autopct='%1.1f%%', colors=['lightcoral','green','red','green','royalblue' ],
        startangle=180, explode=(0.05,0.05,0.05,0.05,0.05))
plt.title('Cars Details')
plt.show()
```

```python
ipl=df['batting_team'].value_counts().head(5)
ipl
```

```python
plt.Figure(figsize=(8,6))
sns.boxplot(x='batting_team',y='total_runs',data=df, palette='Set2')
plt.title('Car Details ')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='match_id',y='total_runs',hue='inning',
                style='batting_team', data=df,palette='rainbow')
plt.title('run score ')
plt.xlabel('match_id')
plt.ylabel('runs')
plt.legend(title='about ipl')
plt.show()
```

```python
df.info()
```

```python
ipl=df[['match_id', 'inning', 'ball']].corr()
ipl
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(ipl,annot=True, cmap='rainbow',center=0)
plt.title('correlation Heatmap of ipl Dataset')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x='inning', y='total_runs', hue='batting_team', data=df,
               split=True, palette='Set2')
plt.title('analysing')
plt.xlabel('inning')
plt.ylabel('total_runs')
plt.legend(title='teams')
plt.show()
```

```python
sns.kdeplot(data=df['total_runs'].dropna(),color='skyblue',shade='')
plt.title('distribution of total_runs')
plt.xlabel('total_runs')
plt.ylabel('Density')
plt.show()
```

```python
g= sns.FacetGrid(df, col='inning', row='batting_team', margin_titles=True)
g.map(sns.histplot,'total_runs',bins=20,color='teal')
g.fig.suptitle('abi',y=1.03)
plt.show()
```

```python
sns.pairplot(df[['batsman_runs', 'total_runs', 'extra_runs', 'batting_team']], palette='husl',
             plot_kws={'alpha':0.6}, diag_kind='kde')
plt.suptitle('innings')
plt.show()
```

```python
plt.Figure(figsize=(5,5))
sns.swarmplot(x='fuel', y='km_driven', hue='transmission',
              data=df, palette='Set1', dodge=True)
plt.title('transmission Distribution by fuel and km_driven (Swarm Plot)')
plt.xlabel('fuel')
plt.ylabel('km_driven')
plt.legend(title='transmission')
plt.show()
```
